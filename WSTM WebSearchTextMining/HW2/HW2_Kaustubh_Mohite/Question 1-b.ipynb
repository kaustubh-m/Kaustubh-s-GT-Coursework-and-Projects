```python
#Find average rating for each user

listUserAvgRtg = [0]*943

for i in range(0,943):
    sumUserRtg = 0
    cntUserRtg = 0
    for j in range(0,1680):
        if (mxUserItem[i,j] > 0):
            sumUserRtg += mxUserItem[i,j]
            cntUserRtg += 1
            print(i,j)
    print(sumUserRtg, cntUserRtg)
    if(cntUserRtg > 0):
        listUserAvgRtg[i] = sumUserRtg / cntUserRtg
    else:
        listUserAvgRtg[i] = 0
```

```python
#function to return set of users that have rated items i1 and i2, both
def retUi12(i1, i2):
    listUi1 = []
    listUi2 = []
    for j in range(0,943):
        if (mxUserItem[j, i1] > 0):
            listUi1.append(j+1)
        if (mxUserItem[j, i2] > 0):
            listUi2.append(j+1)
    listUi12 = [val for val in listUi1 if val in listUi2]
    #for tempUser in listUi12:
    #    for i in range():
    #        mxUserItem[listUser.index(tempUser), ]
    return (listUi12)

#function to calculate similarity of items i1 and i2
def retSim(i1, i2):
    listUi12 = retUi12(i1, i2)
    Nr = 0
    Dr1 = Dr2 = 0
    for tempUser in listUi12:
        Nr += (mxUserItem[listUser.index(tempUser), listItem.index(i1)] - listUserAvgRtg[tempUser-1]) * (mxUserItem[listUser.index(tempUser), listItem.index(i2)] - listUserAvgRtg[tempUser-1])
        Dr1 += (mxUserItem[listUser.index(tempUser), listItem.index(i1)] - listUserAvgRtg[tempUser-1])**2 
        Dr2 += (mxUserItem[listUser.index(tempUser), listItem.index(i2)] - listUserAvgRtg[tempUser-1])**2
    Dr = math.sqrt(Dr1 * Dr2)
    if(Dr == 0):
        sim = 1
    else:
        sim = Nr / Dr
    return (sim)
        


    
        
```
